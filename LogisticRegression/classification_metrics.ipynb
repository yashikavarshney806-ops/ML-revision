```python
df=df.drop_duplicates()
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lor=LogisticRegression(max_iter=100)
lor.fit(X_train,y_train)
```

```python
pred=lor.predict(X_test)
```

```python
print(precision_score(y_test,pred))
print(recall_score(y_test,pred))
print(precision_score(y_test,pred,average=None))
print(recall_score(y_test,pred,average=None))
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```

```python
tree=DecisionTreeClassifier(max_depth=2)
tree.fit(X_train,y_train)
```

```python
y_pred=tree.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(precision_score(y_test,y_pred,average=None))
print(recall_score(y_test,y_pred,average=None))
```

### Multi-class classification

```python
df=pd.read_csv("Datasets/iris.csv")
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=1)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python

clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,3)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,3)))
```

```python
print("Logistic precision : ",precision_score(y_test,y_pred1,average='macro'))
print("Logistic recall : ",recall_score(y_test,y_pred1,average='macro'))
print("Decision precision : ",precision_score(y_test,y_pred2,average='macro'))
print("Decision recall : ",recall_score(y_test,y_pred2,average='macro'))
```

```python
print("Classes learned:", clf1.classes_)
print("Coefficients shape:", clf1.coef_.shape)
print("Intercepts shape:", clf1.intercept_.shape)
```
