```python
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

# Apply Lasso (alpha is lambda)
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

print("Coefficients:", lasso.coef_)
print("Intercept:", lasso.intercept_)
print("R2 Score:", r2_score(y_test, lasso.predict(X_test)))
```

```python
#  Critical Insights from Visual Trends
# Sparsity & Feature Selection: This is the most crucial property of Lasso . As lambda increases, it doesn't just make coefficients small—it forces less important coefficients to become exactly 0 .

# High-Dimensional Data: If you have 100 features but only 5 are actually causing the output, Lasso will automatically select those 5 and kill the remaining 95 features (by making their weights 0) . This is why Lasso serves as an in-built Feature Selection technique .

# Impact on Coefficients: Coefficients with higher magnitude are penalized drastically and drop to zero very fast as lambda grows .
```
